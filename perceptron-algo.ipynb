```python
import numpy as np 
from matplotlib import pyplot as plt 
%matplotlib inline 
```

### Goals 

We want to train a perceptron here by iterating through all the points and update the linear separator by using a learning rate 

### Steps 

1. Randomly initialize and label points
2. Randomly choose starting parameters for a line 
3. Iterate through all points and adjust the parameters 
4. Profit! 

1. Randomly initialize and label points

2. Randomly choose starting parameters for a line 

```python
red_points = np.random.random(size=(1000,2)) * 0.3 + 0.7  # x and y between 0.7 and 1
blue_points = np.random.random(size=(1000,2)) * 0.7  # x and y between 0 and 0.9

fig, ax = plt.subplots()
ax.scatter(red_points[:, 0], red_points[:, 1], c='red')
ax.scatter(blue_points[:, 0], blue_points[:, 1], c='blue')
ax.set_aspect('equal', 'box')

learning_rate = 0.02

x1_coeff = np.random.random() * 2 - 1
x2_coeff = -1 
bias = np.random.random()

# for graphing purposes
x = np.linspace(0, 1, num=1000) 
y = np.clip(x1_coeff * x + bias, 0, 1)

ax.fill_between(x, y, y2=1, where=y<=1,  alpha=0.3, color='r')
ax.fill_between(x, y, y2=0, where=y>=0, alpha=0.3, color='b')

print(x1_coeff, bias)

# update the coefficients 

for point in red_points:

    x1, x2 = point[0], point[1]
    y = x1_coeff * x1 + bias
    if x2 < y:  # misclassified, we should update intercept and slope 
        ## MY MISTAKE IS SOMEWHERE HERE: 
        x1_coeff = x1_coeff - x1 * learning_rate 
        bias = bias - learning_rate 
        x2_coeff = x2_coeff - x2 * learning_rate 

print(x1_coeff, bias)


for point in blue_points:

    x1, x2 = point[0], point[1]
    y = x1_coeff * x1 + bias
    if x2 > y:  # misclassified, we should update intercept and slope 
        x1_coeff = x1_coeff + x1 * learning_rate 
        bias = bias + 2 * learning_rate 
        x2_coeff = x2_coeff + x2 * learning_rate 

print(x1_coeff, bias)
```
